** Задача 3 (5%) **
Докажите, что the k-means всегда сходится.

** Доказательство: **
Докажем, что сумма квадратов расстояний до ближайшего кластера всегда убывает, а из этого уже будет очевидно исходное утверждение. Пусть $A_i$ - наша выборка. X - центр масс, Y - какая-то другая точка, тогда:

$$\sum \overline{YA_i}^2 = \sum (\overline{YA_i},\overline{YA_i}) = \sum (\overline{YX} + \overline{XA_i},\overline{YX} + \overline{XA_i}) = \sum \overline{YX}^2 + \overline{XA_i}^2 + 2 (\overline{YX}, \overline{XA_i}) = n\overline{YX}^2 + \sum \overline{XA_i}^2 + 2 (\overline{YX}, \sum \overline{XA_i}) $$ 
Последняя сумма в скалярном произведении равна нулю, так как Х - центр масс
$$n\overline{YX}^2 + \sum \overline{XA_i}^2 > \sum \overline{XA_i}^2$$

Итого, пусть n_clusters = k, размер выборки - n. Всевозможных кластеризаций выборки - $k^n$. Если на двух последовательных итерациях получено одно и то же разбиение, то значение функционала меньше, чем прежде, поэтому циклов не будет и разбиение на кластера не будут меняться. Иначе - мы будем получать всегда разные разбиения, но так как их конечное число, то алгоритм закончится.

**Задание:** построейте SVD декомпозицию <a href='http://wallpaperswide.com/under_a_yellow_tree-wallpapers.html'> картинки</a> и сожмите её в 2,4,16 и 128 раз (идею см. здесь: http://timbaumann.info/svd-image-compression-demo/).
Отобразите полученные картинки.

```python
X = imread("tree.jpg")
X.shape
```

```python
def pack(X, n_comp = 8):
    X = X.transpose(2, 0, 1)
    udv_tr_3 = []
    for channel in range(3):
        A = X[channel]
        u, d, v = svd(A, full_matrices=False)
        # print(u.shape, d.shape, v.shape)
        u_tr = u[:, :n_comp]
        d_tr = d[:n_comp]
        v_tr = v[:n_comp, :]
        # print(u_tr.shape, d_tr.shape, v_tr.shape)
        udv_tr_3 += [(u_tr, d_tr, v_tr)]
    return udv_tr_3
```

```python
def unpack(udv_tr_3):
    X_tr = []
    for channel in range(3):
        u_tr, d_tr, v_tr = udv_tr_3[channel]
        A_tr = u_tr @ np.diag(d_tr) @ v_tr
        X_tr.append(A_tr)
    X_tr = np.stack(X_tr)
    return X_tr.transpose(1, 2, 0).astype(int)
```

```python
def demo(X, n_comp=8):
    #print(X.shape)
    uvd_tr_3 = pack(X, n_comp)
    size_tr = 0
    for channels in range(3):
        for i in range(3):
            size_tr += np.prod(uvd_tr_3[channels][i].shape)
            # print(np.prod(uvd_tr_3[channels][i].shape))
    # print(size_tr, 3 * n_comp * (1 + np.sum(X.shape)))
    size_real = np.prod(X.shape)
    print("Compression : {}".format(size_real / size_tr))
    plt.imshow(unpack(uvd_tr_3))
    plt.show()
```

```python
demo(X, 8)
```

```python
def ratio_to_ncomp(h, w, ratio):
    return (h * w) / (1 + h + w) / ratio
```

```python
ratios = [2, 4, 16, 128]
ncomps = np.array([ratio_to_ncomp(*X.shape[:2], ratio=ratio) for ratio in ratios]).astype(int)
ncomps
```

```python
for ncomp in ncomps:
    demo(X, ncomp)
```

Now we are going to build a cool model, named autoencoder. The aim is simple: **encode** the data to a lower dimentional representation. Why? Well, if we can **decode** this representation back to original data with "small" reconstuction loss then we can store only compressed representation saving memory. But the most important thing is -- we can reuse trained autoencoder for classification.

Now implement an autoencoder:

Build it such that dimetionality inside autoencoder changes like that: 

$$784 \text{ (data)} -> 512 -> 256 -> 128 -> 30 -> 128 -> 256 -> 512 -> 784$$

Use **MSECriterion** to score the reconstruction.

You may train it for 9 epochs with batch size = 256, initial lr = 0.1 droping by a factor of 2 every 3 epochs. The reconstruction loss should be about 6.0 and visual quality decent already.
Do not spend time on changing architecture, they are more or less the same. 

```python
import os
import numpy as np
from sklearn.datasets import fetch_mldata
import matplotlib.pyplot as plt

print('zuzuuz')
mnist = fetch_mldata("mnist-original")
X, y = mnist.data / 255.0, mnist.target
np.savez('mnist.npz', X=X, y=y)
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder()
y = OHE.fit_transform(y.reshape(-1, 1))
y = y.toarray()
```

```python
y
```

```python
import tensorflow as tf
```

```python
learning_rate = 0.001
inputs_ = tf.placeholder(tf.float32, (None, 28, 28, 1), name='inputs')
targets_ = tf.placeholder(tf.float32, (None, 28, 28, 1), name='targets')


f = tf.contrib.layers.flatten(inputs_)

e1 = tf.contrib.layers.fully_connected(f, 512, activation_fn=tf.nn.relu)
e2 = tf.contrib.layers.fully_connected(e1, 256, activation_fn=tf.nn.relu)
e3 = tf.contrib.layers.fully_connected(e2, 128, activation_fn=tf.nn.relu)
e4 = tf.contrib.layers.fully_connected(e3, 30, activation_fn=tf.nn.relu)

d1= tf.contrib.layers.fully_connected(e4, 30, activation_fn=tf.nn.relu)
d2 = tf.contrib.layers.fully_connected(d1, 128, activation_fn=tf.nn.relu)
d3 = tf.contrib.layers.fully_connected(d2, 256, activation_fn=tf.nn.relu)
d4 = tf.contrib.layers.fully_connected(d3, 512, activation_fn=None)
d5 = tf.contrib.layers.fully_connected(d4, 784, activation_fn=None)

loss = tf.losses.mean_squared_error(
    labels=tf.contrib.layers.flatten(targets_),
    predictions=d5)

opt = tf.train.AdamOptimizer(learning_rate).minimize(loss)
```

```python
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
```

```python
sess = tf.Session()
epochs = 17
batch_size = 64
sess.run(tf.global_variables_initializer())
for e in range(epochs):
    for x_batch, y_batch in get_batches((X, y), batch_size):
        imgs = x_batch.reshape((-1, 28, 28, 1))
        batch_cost, _ = sess.run([loss, opt], feed_dict={inputs_: imgs,
                                                         targets_: imgs})

    print("Epoch: {}/{}...".format(e+1, epochs),
                  "Training loss: {:.4f}".format(batch_cost))
```

Some time ago NNs were a lot poorer and people were struggling to learn deep models. To train a classification net people were training autoencoder first (to train autoencoder people were pretraining single layers with [RBM](https://en.wikipedia.org/wiki/Restricted_Boltzmann_machine)), then substituting the decoder part with classification layer (yeah, they were struggling with training autoencoders a lot, and complex techniques were used at that dark times).
```python
labels_ = tf.placeholder(tf.float32, (None, 10), name='loabels')
cl = tf.contrib.layers.fully_connected(e4, 10, activation_fn=None)
sfmx = tf.losses.softmax_cross_entropy(logits=cl, onehot_labels=labels_)
optcl = tf.train.AdamOptimizer(learning_rate).minimize(sfmx)
```

```python
sess = tf.Session()
epochs = 17
batch_size = 64
sess.run(tf.global_variables_initializer())
for e in range(epochs):
    for x_batch, y_batch in get_batches((X, y), batch_size):
        imgs = x_batch.reshape((-1, 28, 28, 1))
        batch_cost, _ = sess.run([sfmx, optcl], feed_dict={inputs_: imgs,
                                                         labels_: y_batch})

    print("Epoch: {}/{}...".format(e+1, epochs),
                  "Training loss: {:.4f}".format(batch_cost))
```
